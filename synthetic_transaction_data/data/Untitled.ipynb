```python
import pandas as pd
import os 
```

```python
xlsx_files = [f for f in os.listdir("./") if f.endswith('.xlsx')]
```

```python
clients = []
all_df = pd.DataFrame()
for i, file in enumerate(xlsx_files):
    client_name = file[:-5].replace("_", " ").lower().title()
    df = pd.read_excel(file, index_col=0)
    df.insert(0, "Client_Key", i+1)
    df.insert(1, "Client_Name", client_name)
    all_df = pd.concat([all_df, df], axis=0, ignore_index=True)
    clients.append(df)

all_df = all_df.reset_index().rename(columns={"index": "Transaction_ID"})
all_df["Transaction_ID"] = all_df["Transaction_ID"] +1
all_df.to_excel("./ALL_CLIENTS.xlsx", index=False)
```
